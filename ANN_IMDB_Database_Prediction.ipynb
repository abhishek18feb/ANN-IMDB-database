```python
import pandas as pd
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import load_model
from sklearn.preprocessing import StandardScaler, LabelEncoder
import pickle
import numpy as np
```

```python
### Load the trained model, scaler pickle, onehot
model = load_model('/content/model.h5')

with open('tokenizer.pkl', 'rb') as file:
    tokenizer = pickle.load(file)
```

```python
# Example input data
input_data = {
    "review": """
    Bajrangi Bhaijaan is a heart-warming film that beautifully portrays the power of kindness and humanity. Salman Khan delivers one of his most sincere and emotionally grounded performances. The bond between Pawan and Munni is so pure that it brings tears to your eyes. The story flows smoothly with a mix of humor, emotions, and strong messages. The India-Pakistan theme is handled with maturity and positivity. It's a movie that fills your heart with hope and reminds you that compassion still exists in the world.
    """
    }
```

```python
input_data
```

```python
review_df = pd.DataFrame([input_data])
review_df
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
review_df = pd.DataFrame([input_data])
review_seq = tokenizer.texts_to_sequences(review_df['review'])
review_pad = pad_sequences(review_seq, maxlen=200)
print(review_pad)
```

```python
## Prediciton Elegibility for Loan
prediction = model.predict(review_pad)
prediction
```

```python
prediction_prob = prediction[0][0]
prediction_prob
```

```python
if prediction_prob > 0.45:
    print('Postitive Review Comment.')
else:
    print('Negative Review Comment.')
```
